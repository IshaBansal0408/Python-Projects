# Python Plot: Create Animated Plots in Python

### 1. Importing Some Basic Modules

1. Numpy, pandas and matplotlib
2. matplotlib.animation : To provide animation 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.style.use('seaborn')
```

```python
import random
from itertools import count

x=[]
y=[]

i = count()

def animate(j):
    x.append(next(i))
    y.append(random.randint(0, 10))

animation_1 = animation.FuncAnimation(plt.gcf(),animate,interval=1000)

plt.tight_layout()
plt.show()
```

Need to convert it to video format

```python
x=[]
y=[]
i = count()
def animate(j):
    x.append(next(i))
    y.append(random.randint(0, 10))
    plt.cla()
    plt.plot(x,y)
animation_1 = animation.FuncAnimation(plt.gcf(),animate,interval=100)

video_1 = animation_1.to_html5_video()

from IPython import display

html_code_1 = display.HTML(video_1)
display.display(html_code_1)

plt.tight_layout()
plt.show()
```

```python
x1=[]
y1=[]
i1 = count()
def animate1(j):
    t=next(i1)
    x1.append(2*t)
    y1.append(np.sin(t))
    plt.cla()
    plt.plot(x1,y1)
animation_2 = animation.FuncAnimation(plt.gcf(),animate1,interval=50)
video_2 = animation_2.to_html5_video()
html_code_2 = display.HTML(video_2)
display.display(html_code_2)
plt.tight_layout()
plt.show()
```
