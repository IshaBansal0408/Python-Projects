# Plotting 3-Dimensional Plots in Python

### 1. Importing numpy, matplotlib and 3d plotting toolkit for 3D plotting

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
plt.style.use('seaborn')
```

### 2. Creating a blank 3D plot

```python
# Create a figure
fig_1=plt.figure()

# Creating 3d axes
axes_1=plt.axes(projection='3d')
```

### 3. Creating our first simple 3D plot

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

# Creating sample x points and y points
x = np.arange(-5,50,0.05)
y = x
# Creating a z function : in this case a sin function
z1= np.sin(x)
z2=np.cos(x)

plt.title("Sine and Cosine waves - 3D")
axes_2.scatter3D(x,y,z1,color="red",label="Sin_wave")
axes_2.scatter3D(x,y,z2,label="Cos_wave")
plt.legend()
plt.show()
```

3D Scatter points have transparency effects to give a sense of depth on the plot. It can sometimes make it a little difficult to see point behind the static image but with an interactive can lead to visualization of thoes points as well.

### 4. Creating 3D plot surface

#### Wireframe Plots

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

x = np.arange(-5,5,0.05)
y = x
x,y = np.meshgrid(x,y)
z= x+y

plt.title("Equation: z=x+y")
axes_2.plot_wireframe(x,y,z,cmap="rainbow")
plt.show()
```

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

x = np.arange(-5,5,0.05)
y = x
x,y = np.meshgrid(x,y)
z= x**2+y**2

plt.title("Equation: z=x^2+y^2")
axes_2.plot_wireframe(x,y,z,cmap="rainbow")
plt.show()
```

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

x = np.arange(-5,5,0.05)
y = x
x,y = np.meshgrid(x,y)
z= x+y

plt.title("Equation: z=x+y")
axes_2.plot_surface(x,y,z,cmap="rainbow")
plt.show()
```

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

x = np.arange(-5,5,0.01)
y = x
x,y = np.meshgrid(x,y)
z= x**2+y**2

plt.title("Equation: z=x^2+y^2")
axes_2.plot_surface(x,y,z,cmap="rainbow")
plt.show()
```

### 4. Plotting Coutour Plots

```python
fig_2=plt.figure(figsize=(10,10))
axes_2=plt.axes(projection='3d')

x = np.arange(-5,5,0.05)
y = x
x,y = np.meshgrid(x,y)
z= x+y

plt.title("Equation: z=x+y")
axes_2.contour3D(x,y,z,cmap="rainbow")
plt.show()
```
