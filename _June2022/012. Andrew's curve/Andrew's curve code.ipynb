```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')


df = pd.read_csv(
    'https://raw.github.com/pandas-dev/'
    'pandas/main/pandas/tests/io/data/csv/iris.csv'
)

plt.figure(figsize=(20,15),facecolor="w")
pd.plotting.andrews_curves(df, 'Name')
plt.title("Andrew Curve for Iris Dataset!",fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')

df = pd.read_csv('adult.csv')
df = df.drop(["Education","Marital Status","Occupation",
              "Relationship","Race","Sex","Country","Above/Below 50k"],axis=1)

plt.figure(figsize=(20,15),facecolor="w")
pd.plotting.andrews_curves(df, 'Workclass')
plt.title("Andrew Curve for Adult Dataset!",fontsize=20)
plt.show()
```
