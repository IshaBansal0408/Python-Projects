# ClusterMap in Python

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/mtcars.csv")
df = df.set_index('model')

sns.set(rc={'figure.facecolor':'white'})
sns.clustermap(df)
plt.title("Default Clustermap for the dataset",fontsize=20)
plt.show()
```

```python
sns.clustermap(df, standard_scale=1)
plt.title("Clustermap after Normalization for the dataset",fontsize=20)
plt.show()
```

```python
sns.clustermap(df, standard_scale=1,cmap="coolwarm")
plt.title("Clustermap Changed colormap for the dataset",fontsize=20)
plt.show()
```

```python
sns.clustermap(df, standard_scale=1,cmap="coolwarm",annot=True)
plt.title("Clustermap Apply Annotations for the dataset",fontsize=20)
plt.show()
```

```python
sns.clustermap(df, standard_scale=1,cmap="rainbow",annot=True,method="single")
plt.title("Clustermap with Single as the method for the dataset",fontsize=20)
plt.show()
```
