Let's consider a situation where you have multiple tasks to complete in `24 hours` for `7 days` continuously. The tasks involve: Reading Book, Working on the app, Yoga + Workout and taking some time out to sleep.


```python
import matplotlib.pyplot as plt

days            =  [1,2,3,4,5,6,7]

sleeping        =  [7,8,6,5,7,8,9]
reading_books   =  [4,6,4,7,4,5,5]
yoga_workout    =  [3,2,3,4,2,4,2]
working_on_app  =  [10,8,11,8,11,7,8]
```

So, our `X` axis will consist of the days variable, which is a week implying the values to be `1,2,3,4,5,6 and 7`. And the rest of the values are the hours given to the activies for each day and the distribution of it.

Look at the code and the resulting plot below.

```python
plt.figure(figsize=(10,5),facecolor="w")
plt.stackplot(days,sleeping,reading_books,
              yoga_workout,working_on_app,
              colors=['m','c','r','k'])
plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Stack Plot for the Tasks distribution over the week')
plt.show()
```

The problem here in the plot ist that we don't really know which color represent which activity without having prior information about in from the code.

Another problem with the plot is that we don't have the labels for the datavalues which makes it difficult to understand the time distribution from the plot. 

Let's try to understand the data represrented in the plot through the code below. Look at the output as well.

```python
plt.figure(figsize=(10,5),facecolor="w")

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='reading_books', linewidth=5)
plt.plot([],[],color='r', label='yoga_workout', linewidth=5)
plt.plot([],[],color='k', label='working_on_app', linewidth=5)

plt.stackplot(days,sleeping,reading_books,
              yoga_workout,working_on_app,
              colors=['m','c','r','k'])

plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Stack Plot for the Tasks distribution over the week')
plt.legend()
plt.show()
```

There is a slight issue here. The issue is that the base of the stackplot is kind of merging with the axis. To fix that let's add a parameter `baseline` to `sym` making the plot symmetrical.

```python
plt.figure(figsize=(10,5),facecolor="w")

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='reading_books', linewidth=5)
plt.plot([],[],color='r', label='yoga_workout', linewidth=5)
plt.plot([],[],color='k', label='working_on_app', linewidth=5)

plt.stackplot(days,sleeping,reading_books,
              yoga_workout,working_on_app,
              colors=['m','c','r','k'],baseline="sym")

plt.xlabel('X data')
plt.ylabel('Y data')
plt.title('Stack Plot for the Tasks distribution over the week')
plt.legend()
plt.show()
```
