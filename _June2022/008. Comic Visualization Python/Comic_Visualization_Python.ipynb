`xkcd` is a webcomic created in 2005 by American author Randall Munroe. “A webcomic of romance, sarcasm, math, and language.” xkcd is an online Comic Strip that airs a new comic every week. It is one of the more popular comics. Most of them are related to math, computer programming or the internet.



## Load Libraries and Data

We will start off my importing the necesaary librarines and import the data into the program as well. Here we load the `tips` dataset and print the first `five` rows of the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset

tips = load_dataset("tips")
tips.head()
```

## Example 1 - Bar Plots

A bar graph is a graphical representation of data in which we can highlight the category with particular shapes like a rectangle. The length and heights of the bar chart represent the data distributed in the dataset. 

We can plot the basic bar chart using the `sns.countplot` which will plot the counts of unique values of a column in form of a bar plot.

```python
plt.figure(facecolor="W",figsize=(10,5))
plt.title("Bar Plot - Normal")
sns.countplot(tips['sex'])
plt.show()
```

To make the plot comic all we have to go is add `plt.xkcd` function before the whole code.

```python
plt.xkcd()
plt.figure(facecolor="W",figsize=(10,5))
plt.title("Box Plot - Comic")
sns.countplot(tips['sex'])
plt.show()
```

## Example 2 : Box Plot

Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set.

We can plot the boxplot using the `sns.boxplot` using the code below.

```python
plt.figure(facecolor="W",figsize=(10,5))
plt.title("BoxPlot - Normal")
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = tips);
plt.show()
```

The plot can be made comic with the help of the `plt.xkcd` function as displayed in the code below.

```python
plt.xkcd()
plt.figure(facecolor="W",figsize=(10,5))
plt.title("BoxPlot - Comic")
sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = tips);
plt.show()
```

## Example 3 - Sine Wave 

In this example we will be plotting a sine wave plot by generating the values using the `linspace` and `sin` functions as shown in the code below.

```python
plt.figure(facecolor="W",figsize=(10,5))
plt.plot(np.sin(np.linspace(0, 10)))
plt.title('A simple Sine Wave - Normal')
plt.show()
```

To make the plot comic, we make use of the `plt.xkcd` function using the code below.

```python
plt.xkcd()
plt.figure(facecolor="W",figsize=(10,5))
plt.plot(np.sin(np.linspace(0, 10)))
plt.title('A simple Sine Wave - Comic')
plt.show()
```

## Example 4 - Pie Chart

Matplotlib supports pie charts using the pie() function. You can use the method plt.pie() to create a plot. The code below creates a pie chart:

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.figure(facecolor="W",figsize=(10,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Pie Chart - Normal")
plt.show()
```

We can convert the plot to comic using the `plt.xkcd` function using the code below.

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
plt.xkcd()
plt.figure(facecolor="W",figsize=(10,5))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Pie Chart - Comic")
plt.show()
```
