```python
from mpl_toolkits.mplot3d import axes3d 
from matplotlib import pyplot 
```

```python
fig = pyplot.figure(figsize=(10,10)) 

wf = fig.add_subplot(111, projection='3d') 

x, y, z = axes3d.get_test_data(0.07)
```

```python
fig = pyplot.figure(figsize=(10,10)) 

wf = fig.add_subplot(111, projection='3d') 

x, y, z = axes3d.get_test_data(0.07)

wf.plot_wireframe(x,y,z, rstride=2, cstride=2, color='orchid')

wf.set_title('3D wireframe plot Example - 1') 
```

```python
import numpy 

a = numpy.linspace(-10, 15, 25) 
b = numpy.linspace(-15, 10, 35) 
x, y = numpy.meshgrid(a, b) 
z = numpy.cos(numpy.sqrt(x**2 + y**2)) 

fig = pyplot.figure(figsize=(30,20)) 
wf = pyplot.axes(projection ='3d') 
wf.plot_wireframe(x, y, z, color ='green') 

wf.set_title('3D wireframe plot Example - 2') 
pyplot.show() 
```
