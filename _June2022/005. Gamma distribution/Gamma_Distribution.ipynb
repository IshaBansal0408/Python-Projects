```python
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt
```

```python
x = np.linspace (0, 40, 100) 
y = stats.gamma.pdf(x, a=5, scale=3)
color = np.arange(100)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x,y,c=color)
plt.title("Scatter plot for Gaussian Distribution",fontsize=20,color="magenta")
plt.show()
```

```python
x = np.linspace (0, 40, 100) 

y1 = stats.gamma.pdf(x, a=5, scale=3)
y2 = stats.gamma.pdf(x, a=5, scale=2)
y3 = stats.gamma.pdf(x, a=5, scale=7)

plt.figure(figsize=(15,10))
plt.scatter(x,y1,color="red",label="Scale = 3")
plt.scatter(x,y2,color="green",label="Scale = 2")
plt.scatter(x,y3,color="magenta",label="Scale = 7")

plt.legend()
plt.title("Multiple Scatter plot for Gaussian Distribution",fontsize=20,color="brown")
plt.show()
```
