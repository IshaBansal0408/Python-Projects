```python
import numpy as np

def fun(x,y):
   return np.sin(x) ** 40 + np.cos(20 + y * x) * np.cos(x)

x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 40)

X, Y = np.meshgrid(x, y)
Z = fun(X, Y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.contour(X, Y, Z, colors='black')
plt.title("Standard Contour Plot",fontsize=20,color="red")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.contour(X, Y, Z, cmap='rainbow')
plt.title("ColorMap Added Contour Plot",fontsize=20,color="Blue")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.contourf(X, Y, Z, cmap='rainbow')
plt.title("Contourf Plot",fontsize=20,color="Blue")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.contourf(X, Y, Z, cmap='rainbow')
plt.title("Contourf Plot - ColorBar Added",fontsize=20,color="Blue")
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
plt.title("Contourf Plot - Better Rendering",fontsize=20,color="Magenta")
plt.show()
```

```python
plt.figure(figsize=(10,10))

contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)

plt.colorbar()

plt.title("Combining Contour and Contourf Plot",fontsize=20,color="Green")
plt.show()
```
