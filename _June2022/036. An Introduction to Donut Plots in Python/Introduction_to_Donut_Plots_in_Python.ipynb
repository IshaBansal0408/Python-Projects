```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(10, 100, 10)
```

```python
plt.style.use('seaborn')

plt.figure(figsize=(5,5),facecolor="w")
plt.pie(data)
plt.title("Simple Pie Chart For the dataset",fontsize=15)
plt.savefig("PieChart_Donut.png")

plt.show()
```

```python
plt.figure(figsize=(5,5),facecolor="w")

plt.pie(data)
circle = plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title("Simple Donut Chart For the dataset",fontsize=15)
plt.savefig("DonutChart_Plain.png")

plt.show()
```

```python
plt.figure(figsize=(5,5),facecolor="w")

plt.pie(data,labels=['.Net','C++','C','Java','Kotlin',
                     'Python','Nodejs','Javascript','AngularJs','Django'])
circle = plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title("Donut Chart with Labels",fontsize=15)
plt.savefig("DonutChart_With_Labels.png")

plt.show()
```

```python
total = sum(data)
data_per = data/total*100
labels=['.Net','C++','C','Java','Kotlin','Python',
        'Nodejs','Javascript','AngularJs','Django']
updated_labels = [labels[i] + " => " + str(round(data_per[i],2)) + "%" for i in range(len(list(data_per)))]

plt.figure(figsize=(5,5),facecolor="w")

plt.pie(data,labels=updated_labels)
circle = plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
plt.title("Donut Chart with Percentages",fontsize=15)
plt.savefig("DonutChart_With_Percentage.png")

plt.show()
```
