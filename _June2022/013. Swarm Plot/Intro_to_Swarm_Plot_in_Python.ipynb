```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('white')
plt.style.use('seaborn')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
cars = sns.load_dataset('mpg').dropna()
print("Shape of Data : ",cars.shape)
cars.head()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
plt.figure(figsize=(10,5),facecolor="w")
plt.title("Simple Swarm Plot")
sns.swarmplot(cars.horsepower)
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,5),facecolor="w")
plt.title("Swarm Plot - x and y data split")
sns.swarmplot(x=cars.origin, y=cars.horsepower)
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,5),facecolor="w")
sns.swarmplot(x='origin', y='horsepower', hue='cylinders',
               data=cars)
plt.title("Swarm Plot - x and y value + Hue")
plt.show()
```
